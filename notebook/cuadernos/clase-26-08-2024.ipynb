```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt

df_iris = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv", sep=",")

df_iris
```

```python
#Realizar filtros usando condiciones.

df_filtrado = df_iris[df_iris["species"]=="virginica"]
df_filtrado.head()
```

```python
df_filtrado = df_iris[(df_iris["sepal_length"]>5.0)&(df_iris["species"]=="virginica")]
df_filtrado.head()
```

```python
df_filtrado1 = df_iris[df_iris["species"].isin(["setosa","versicolor"])]
df_filtrado1
```

```python
dfparacorelacion = df_iris[["sepal_width","sepal_length","petal_length","petal_width"]]


corelacion_variables = dfparacorelacion.corr()

print(corelacion_variables)
```

```python
plt.figure(figsize=(8,6))
plt.imshow(corelacion_variables,cmap='coolwarm', interpolation='none')
plt.colorbar()


plt.xticks(np.arange(len(corelacion_variables.columns)),corelacion_variables.columns)
plt.xticks(np.arange(len(corelacion_variables.columns)),corelacion_variables.columns)


plt.show()
```
