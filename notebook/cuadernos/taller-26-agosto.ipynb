```python
datosparacorrelacion = df_1[['area', 'geoname_id', 'numeric', 'population']]

correlacion = datosparacorrelacion.corr()

type(correlacion)
#print(correlacion)
```

```python
plt.figure(figsize=(8,6))
plt.imshow(correlacion,cmap='coolwarm', interpolation='none')
plt.colorbar()

##etiqueta de los ejes
plt.xticks(np.arange(len(correlacion.columns)),correlacion.columns, rotation=90)
plt.yticks(np.arange(len(correlacion.columns)),correlacion.columns)

for i in range(len(correlacion.columns)):
    for j in range(len(correlacion.columns)):
        plt.text(j,i,f"{correlacion.iloc[i,j]:.2f}",ha="center",va="center", color="white")

plt.title("Matriz de correlaciòn de variables")
plt.tight_layout()
plt.show()
```

### En este caso solo pudimos aplicar correlación con las columnas las cuales el tipo de datos son numericos, dejando al final 4 columnas. En el grafico de la aprte superior se observa que son muy poco correlacionados los datos entre si. Donde se observa mayor correlación es el area con el numero de población, siendo del 41%, es decir, con ese porcentaje no se podria afirmar que un pais con mayor area tiene mayor población.

# Ejercicio 2:

```python
df_2 = pd.read_csv("https://raw.githubusercontent.com/DrueStaples/Population_Growth/master/countries.csv")
df_2
```

### Es una tabla que contiene el pais, el año y la población

```python
df_filtradoespaña = df_2[df_2["country"]=="Spain"]
df_filtradoespaña
```

### Filtramos por solamente españa y lo almacenamos en un data frame

```python
plt.bar(x = df_filtradoespaña["year"], height = df_filtradoespaña["population"] )
plt.show
```

### Podemos graficar en un diagrama de barras el año vs la población de españa; se observa como disminuyó el crecimiento de la poblacion entre 1980 y 1990

```python
df_filtradoargentina = df_2[df_2["country"]=="Argentina"]
df_filtradoargentina
```

### Filtramos por Argentina

```python
plt.bar(df_filtradoargentina["year"], df_filtradoargentina["population"], width=2, label="Argentina", color='red')
plt.bar(df_filtradoespaña["year"], df_filtradoespaña["population"], width=1, label="Spain", color='blue')

plt.xlabel("Año")
plt.ylabel("Población")
plt.title("Población de Argentina y España por año")

# Mostrar leyenda
plt.legend()


plt.show()
```

### Graficamos en un diagrama de barras, los dataframe de España y Brasil para realizar una comparación. Se observa que desde 1950 España ha tenido más población que Argentina, pero esta diferencia ha disminuido con los años

```python
cajabigote = df_2['year'].dropna()

# Dibujamos la gráfica de cajas y bigotes.
plt.boxplot([cajabigote])
plt.show()
```

### Si realizamos una caja y bogites de la columna Year vemos que no se presentan outliers, y la forma de la caja es facil de apreciar

```python
df_2.describe()
```

### Podemos usar .describe() para comprobarlo.
### Bajo el criterio de que será outliers todo aquello que sea mayor a la media mas 2 desviaciones estandaar, entonces seria outliers todo dato de year mayor a 2014.03066. Pero esto no ocurre ya que el dato maximo es 2007.
### Pero esto no ocurre con la columna populacion, cuyo limites en el orden de e+08, es superado por datos maximos de e+09.
### Para comprobar esto realicemos un diagrama de caja y bigotes.

```python
outliers = df_2['population'].dropna()

plt.boxplot([outliers])
plt.show()
```

### La caja es muy pequeña para poder ser vista, además de haber muchos outliers en el grafico. Siendo que la cota maxima es 241917010, filtraremos los registros mayores a este.

```python
datos_outliers = df_2[df_2['population'] > 241917010]
datos_outliers
```

### Aunque sean outliers, por el contexto de los datos no es aconsejable eliminarlo. En los registros se encuentran casos como China e India, cuyos datos de poblacion no deberian considerarse outliers, solamente es propio de la naturaleza del caso.
