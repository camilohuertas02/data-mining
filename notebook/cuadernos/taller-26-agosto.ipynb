```python
datosparacorrelacion = df_1[['area', 'geoname_id', 'numeric', 'population']]

correlacion = datosparacorrelacion.corr()

type(correlacion)
#print(correlacion)
```

```python
plt.figure(figsize=(8,6))
plt.imshow(correlacion,cmap='coolwarm', interpolation='none')
plt.colorbar()

##etiqueta de los ejes
plt.xticks(np.arange(len(correlacion.columns)),correlacion.columns, rotation=90)
plt.yticks(np.arange(len(correlacion.columns)),correlacion.columns)

for i in range(len(correlacion.columns)):
    for j in range(len(correlacion.columns)):
        plt.text(j,i,f"{correlacion.iloc[i,j]:.2f}",ha="center",va="center", color="white")

plt.title("Matriz de correlaciÃ²n de variables")
plt.tight_layout()
plt.show()
```

### En este caso solo pudimos aplicar correlaciÃ³n con las columnas las cuales el tipo de datos son numericos, dejando al final 4 columnas. En el grafico de la aprte superior se observa que son muy poco correlacionados los datos entre si. Donde se observa mayor correlaciÃ³n es el area con el numero de poblaciÃ³n, siendo del 41%, es decir, con ese porcentaje no se podria afirmar que un pais con mayor area tiene mayor poblaciÃ³n.

```python
df_2 = pd.read_csv("https://raw.githubusercontent.com/DrueStaples/Population_Growth/master/countries.csv")
df_2
```

### Es una tabla que contiene el pais, el aÃ±o y la poblaciÃ³n

```python
df_filtradoespaÃ±a = df_2[df_2["country"]=="Spain"]
df_filtradoespaÃ±a
```

### Filtramos por solamente espaÃ±a y lo almacenamos en un data frame

```python
plt.bar(x = df_filtradoespaÃ±a["year"], height = df_filtradoespaÃ±a["population"] )
plt.show
```

### Podemos graficar en un diagrama de barras el aÃ±o vs la poblaciÃ³n de espaÃ±a; se observa como disminuyÃ³ el crecimiento de la poblacion entre 1980 y 1990

```python
df_filtradoargentina = df_2[df_2["country"]=="Argentina"]
df_filtradoargentina
```

### Filtramos por Argentina

```python
plt.bar(df_filtradoargentina["year"], df_filtradoargentina["population"], width=2, label="Argentina", color='red')
plt.bar(df_filtradoespaÃ±a["year"], df_filtradoespaÃ±a["population"], width=1, label="Spain", color='blue')

plt.xlabel("AÃ±o")
plt.ylabel("PoblaciÃ³n")
plt.title("PoblaciÃ³n de Argentina y EspaÃ±a por aÃ±o")

# Mostrar leyenda
plt.legend()


plt.show()
```

### Graficamos en un diagrama de barras, los dataframe de EspaÃ±a y Brasil para realizar una comparaciÃ³n. Se observa que desde 1950 EspaÃ±a ha tenido mÃ¡s poblaciÃ³n que Argentina, pero esta diferencia ha disminuido con los aÃ±os

```python
cajabigote = df_2['year'].dropna()

# Dibujamos la grÃ¡fica de cajas y bigotes.
plt.boxplot([cajabigote])
plt.show()
```

### Si realizamos una caja y bogites de la columna Year vemos que no se presentan outliers, y la forma de la caja es facil de apreciar

### Podemos usar .describe() para comprobarlo.
### Bajo el criterio de que serÃ¡ outliers todo aquello que sea mayor a la media mas 2 desviaciones estandaar, entonces seria outliers todo dato de year mayor a 2014.03066. Pero esto no ocurre ya que el dato maximo es 2007.
