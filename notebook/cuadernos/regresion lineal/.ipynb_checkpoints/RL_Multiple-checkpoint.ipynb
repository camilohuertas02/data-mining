# Regresión Lineal Múltiple

Este notebook muestra cómo realizar una regresión lineal múltiple utilizando Python y la biblioteca `scikit-learn`. Usaremos un conjunto de datos ficticio para ilustrar el proceso.

```python
# Importar las bibliotecas necesarias
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Generar un conjunto de datos ficticio
np.random.seed(0)
X = np.random.rand(100, 3) * 10  # 100 muestras, 3 características
y = 3 + 2 * X[:, 0] + 1.5 * X[:, 1] + np.random.randn(100) * 2  # Ecuación: y = 3 + 2*x1 + 1.5*x2 + ruido

# Crear un DataFrame
df = pd.DataFrame(X, columns=['Característica 1', 'Característica 2', 'Característica 3'])
df['Objetivo'] = y
df.head()  # Mostrar las primeras filas del DataFrame
```

```python
# Gráfica de dispersión para las características
plt.figure(figsize=(12, 6))

# Gráfica para la Característica 1
plt.subplot(1, 3, 1)
plt.scatter(df['Característica 1'], df['Objetivo'], color='blue', alpha=0.5)
plt.xlabel('Característica 1')
plt.ylabel('Objetivo')
plt.title('Característica 1 vs Objetivo')

# Gráfica para la Característica 2
plt.subplot(1, 3, 2)
plt.scatter(df['Característica 2'], df['Objetivo'], color='orange', alpha=0.5)
plt.xlabel('Característica 2')
plt.ylabel('Objetivo')
plt.title('Característica 2 vs Objetivo')

# Gráfica para la Característica 3
plt.subplot(1, 3, 3)
plt.scatter(df['Característica 3'], df['Objetivo'], color='orange', alpha=0.5)
plt.xlabel('Característica 3')
plt.ylabel('Objetivo')
plt.title('Característica 3 vs Objetivo')

plt.tight_layout()
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

# Crear el gráfico 3D
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Graficar las características
ax.scatter(df['Característica 1'], df['Característica 2'], df['Característica 3'], c=df['Objetivo'], cmap='viridis', alpha=0.7)

# Etiquetas de los ejes
ax.set_xlabel('Característica 1')
ax.set_ylabel('Característica 2')
ax.set_zlabel('Característica 3')
ax.set_title('Gráfico de Dispersión 3D de las Características')

plt.show()
```

## División de los datos
Dividimos los datos en conjuntos de entrenamiento y prueba.

```python
# Dividir los datos en conjunto de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

## Entrenamiento del modelo
Entrenamos el modelo de regresión lineal múltiple utilizando el conjunto de entrenamiento.

```python
# Crear el modelo de regresión lineal
model = LinearRegression()

# Entrenar el modelo
model.fit(X_train, y_train)
```

## Predicciones
Realizamos predicciones sobre el conjunto de prueba.

```python
# Realizar predicciones
y_pred = model.predict(X_test)
```

## Evaluación del modelo
Evaluamos el rendimiento del modelo utilizando el error cuadrático medio y el coeficiente de determinación (R^2).

```python
# Evaluar el modelo
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'Error cuadrático medio: {mse}')
print(f'R^2: {r2}')
```

### Supuestos

```python
# Calcular los residuos
residuos = y_test - y_pred

# Gráfica de residuos
plt.figure(figsize=(8, 6))
plt.scatter(y_pred, residuos, color='green', alpha=0.5)
plt.axhline(0, color='red', linestyle='--')
plt.xlabel('Predicciones')
plt.ylabel('Residuos')
plt.title('Gráfica de Residuos')
plt.show()
```

## Visualización de los resultados
Vamos a visualizar las predicciones frente a los valores reales.

```python
# Visualizar los resultados
plt.scatter(y_test, y_pred, color='blue')
plt.xlabel('Valores reales')
plt.ylabel('Predicciones')
plt.title('Valores Reales vs Predicciones. Tipo Q-Q Line') #OJO NO ES EXACTAMENTE UN Q-Q Line
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red')  # Línea de referencia
plt.show()
```
