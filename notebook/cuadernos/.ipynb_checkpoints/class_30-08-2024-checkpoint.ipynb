## Procesamiento de datos y  predicci√≥n

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
data = datasets.load_iris()
type(data)
```

```python
df = pd.DataFrame(data.data, columns=data.feature_names)
df["especie"] = data.target
type(df)
```

```python
correlacion = df.corr()
correlacion
```

```python
#fig, ax=plt.subplots()
#cax = ax.natshow(correlacion, cmap="coolwarm")




plt.figure(figsize=(8,6))
plt.imshow(correlacion,cmap='coolwarm', interpolation='none')
plt.colorbar()

##etiqueta de los ejes
plt.xticks(np.arange(len(correlacion.columns)),correlacion.columns, rotation=90)
plt.yticks(np.arange(len(correlacion.columns)),correlacion.columns)

for i in range(len(correlacion.columns)):
    for j in range(len(correlacion.columns)):
        plt.text(j,i,f"{correlacion.iloc[i,j]:.2f}",ha="center",va="center", color="white")

plt.title("Matriz de correlaci√≤n de variables")
plt.tight_layout()
plt.show()
```

```python
y_real = df["petal width (cm)"]
x = df["petal length (cm)"]

#def RMSE(x,y_real,m):
#    RMSE=[]
#    for i in m:
#        for dato in x:
#                y_pred=i*dato
#                RMSE.append()


#residuo=(y_real - y_predic)


# determinamos los valores y_real
def RMSE(x,y_real,m):
    y_real_list=[]
    y_pred_list=[]
    residuo_list=[]
    for i in range(149):
        y_real_list.append(y_real[i])
    
    # determinamos los valres y_pred
    
    for dato in x:
        y_pred = m * dato
        y_pred_list.append(y_pred)
    
    for i in range(149):
        residuo = y_real_list[i]-y_pred_list[i]
        residuo_list.append(residuo*residuo)
    
    for i in range(149):
        suma=0
        suma=suma+residuo_list[i]
    
    suma = suma / 150
    rmse = np.sqrt(suma)
    return rmse
```

```python
RMSE_list=[]
m = np.linspace(0.2,0.4,10)

for i in m:
    dato=RMSE(x, y_real, i)
    RMSE_list.append(dato)

RMSE_list
```
