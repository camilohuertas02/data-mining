# Manejo de DataFrame (Filtros)

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
url="https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv"
df=pd.read_csv(url, sep=",")
df.head()
```

```python
#Realizar filtros usando condiciones
df_filtrado=df[(df["sepal_length"]>5.0) | (df["species"]=="virginica")]
df_filtrado.tail()
```

```python
df_filtrado1=df[df["species"].isin(["setosa","versicolor"])]
df_filtrado1
```

```python
df_filtrado2=df[["sepal_length","sepal_width", "petal_length","petal_width"]]
df_filtrado2
```

```python
corelacion_variables=df_filtrado2.corr()
print(corelacion_variables)
```

```python
### Crear un grafico que muestre las correlaciones
plt.figure(figsize=(8,6))
plt.imshow(corelacion_variables,cmap="coolwarm",interpolation="none")
plt.colorbar()

##etiqueta de los ejes
plt.xticks(np.arange(len(corelacion_variables.columns)),corelacion_variables.columns, rotation=90)
plt.yticks(np.arange(len(corelacion_variables.columns)),corelacion_variables.columns)

for i in range(len(corelacion_variables.columns)):
    for j in range(len(corelacion_variables.columns)):
        plt.text(j,i,f"{corelacion_variables.iloc[i,j]:.2f}",ha="center",va="center", color="white")

plt.title("Matriz de corelaciòn de variables")
plt.tight_layout()
plt.show()
```
