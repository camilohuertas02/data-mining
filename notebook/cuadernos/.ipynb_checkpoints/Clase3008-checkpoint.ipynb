# Proceso de Modelado y Predicción

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
data=datasets.load_iris()
type(data)
```

```python
df=pd.DataFrame(data.data, columns=data.feature_names)
df["especie"]=data.target
df
```

```python
cor=df.corr()
type(cor)
```

```python
#Grafica para la matrix de correlación
fig, ax=plt.subplots()
cax=ax.matshow(cor, cmap="coolwarm")
fig.colorbar(cax)
#etiquetas
ax.set_xticks(np.arange(len(cor.columns)))
ax.set_yticks(np.arange(len(cor.columns)))
#nombres
ax.set_xtickslabels(cor.columns)
ax.set_ytickslabels(cor.columns)
#rotar el nombre de ña etiqueta de x
plt.xticks(rotation=45)
plt.show()
```

```python
plt.scatter(df["petal length (cm)"],df["petal width (cm)"])

#Graficar una linea recta.
m=np.linspace(0.2,0.4,10)

for i in m:
    plt.plot(df["petal length (cm)"],i*df["petal length (cm)"],color="r")


plt.title("Grafico de Variables con alta Correlación")
plt.xlabel("Londitud del Petalo")
plt.ylabel("Ancho del Petalo")
plt.show()
```
