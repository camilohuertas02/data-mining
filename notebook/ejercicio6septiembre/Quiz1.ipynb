```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('Datos.xlsx')

X = df['ingresos']
Y = df['consumo']

x = X[:100]  # Tomamos los primeros 100 datos para entrenamiento
y = Y[:100]

w_0 = 0.3  # Pendiente inicial
num_iter = 100  # Número de iteraciones
alpha = 10 ** -5  #learning rate

# Inicialización de la pendiente
w = w_0

# Listas para almacenar los valores de w y de la función de costo
w_act = []
cost_act = [] #no es necesario para el modelo pero nos permite observar la evolución

for i in range(num_iter):
    # Definimos funcion de costo
    J = np.sum(((w * x) - y) ** 2) / len(x)
    cost_act.append(J)
    # Derivada de la función de costo con respecto a w
    dj = 2 * np.sum(((w * x) - y) * x) / len(x)
    # Actualizamos el valor de w
    w -= alpha * dj
    # Guardamos el valor de w
    w_act.append(w)

# Grafica los datos vs el modelo
plt.figure(figsize=(10, 6))

# Datos reales
plt.scatter(X, Y, color='blue', label='Datos reales')

# Targets modelo
plt.plot(X, w * X, color='red', linestyle='-', linewidth=2, label='Modelo entrenado')

plt.title('Datos y Modelo')
plt.xlabel('Ingresos')
plt.ylabel('Consumo')
plt.legend()
plt.grid(True)
plt.show()
```
